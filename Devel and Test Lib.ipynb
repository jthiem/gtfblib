```python
%matplotlib inline
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import gtfblib

timpulse = np.zeros(2**14)
timpulse[100] = 1
```

```python
fbchen = gtfblib.Chen()
iresponse = fbchen.process(timpulse)
for n in range(iresponse.shape[0]):
    plt.magnitude_spectrum(iresponse[n, :], scale='dB', Fs=16000)
```

```python
fbfirc = gtfblib.FIR(complexresponse=True)
iresponse = fbfirc.process(timpulse)
for n in range(iresponse.shape[1]):
    plt.magnitude_spectrum(iresponse[:, n], scale='dB', Fs=16000)
```
